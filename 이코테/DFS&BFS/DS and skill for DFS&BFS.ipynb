<h3>let's look at some of them in real code</h3>

```python
# making the conditions of end-call

def recursive_function(i):
    if i > 100:
        return
    
    print('현재 호출 횟수는:', i)
    i += 1
    recursive_function(i)
    
recursive_function(1)
```

```python
'''
representation of factorial in two ways

-first way: iterative
-second way: recursive
'''

def first_way(n):
    res = 1
    
    for i in range(1, n+1):
        res *= i
        
    return res

print(first_way(7))

def second_way(m):
    # condition of end_call
    if m == 1:
        return 1
    
    res = m * second_way(m-1)
    return res
    
    # other way to represent above code
    # return m * secod_way(m-1)


print(second_way(7))
```
