```python
# way to slove this on book
from collections import deque

n, m = map(int, input().split())
graph = []

for i in range(n):
    graph.append(list(map(int, input())))
    

# i have to know this
# it's skill which is usefull when i try to make code for direction
# it's gonna save 4 directions in graph
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(x, y):
    q = deque()
    q.append((x, y))
    
    while q:
        x, y = q.popleft()
        
        # this is usefull skill!!!
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            # skip when you are over the range
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
                
            # wall = 0 so that ignore!!
            if graph[nx][ny] == 0:
                continue
                
            # when graph[nx][ny] is grater than 1, that's mean you already visited 
            # so in that case, skip
            if graph[nx][ny] == 1:
                # count how many i walk through path
                graph[nx][ny] = graph[x][y] + 1
                q.append((nx, ny))
                
    return graph[n-1][m-1]
print(bfs(0, 0))
```
