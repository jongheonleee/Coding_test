```python
# 답안 예시

n, m = map(int, input().split())

d = [[0] * m for _ in range(n)]

y, x, direction = map(int, input().split())
d[y][x] = 1

arr = []
for i in range(n):
    arr.append(list(map(int, input().split())))
    
dy, dx = [-1, 0, 1, 0], [0, 1, 0, -1]

def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3
        
cnt, turn_time = 1, 0

while 1:
    turn_left()
    ny, nx = y + dy[direction], x + dx[direction]
    
    if d[ny][nx] == 0 and arr[ny][nx] == 0:
        d[ny][nx] = 1
        y, x = ny, nx
        
        cnt += 1
        turn_time = 0
        continue
        
    else:
        turn_time += 1
        
    if turn_time == 4: 
        ny, nx = y - dy[direction], x - dx[direction]
        
        if arr[ny][nx] == 0:
            y, x = ny, nx
            
        else:
            break
            
        turn_time = 0
        
print(cnt)
```
